```python
df3['Memory'] = df3['Memory'].astype(str).replace('\.0','', regex=True)
df3['Memory'] = df3['Memory'].str.replace('GB', '')
df3['Memory'] = df3['Memory'].str.replace('TB', '000')
new = df3['Memory'].str.split('+',n=1, expand = True)

df3['first'] = new[0]
df3['first'] = df3['first'].str.strip()

df3['second'] = new[1]

df3['Layer1HDD'] = df3['first'].apply(lambda x:1 if 'HDD' in x else 0)
df3['Layer1SSD'] = df3['first'].apply(lambda x:1 if 'SSD' in x else 0)
df3['Layer1Hybrid'] = df3['first'].apply(lambda x:1 if 'Hybrid' in x else 0)
df3['Layer1Flash_Storage'] = df3['first'].apply(lambda x:1 if 'Flash_Storage' in x else 0)

df3['first'] = df3['first'].str.replace(r'\D','')
df3['second'].fillna('0',inplace = True)

df3['Layer2HDD'] = df3['first'].apply(lambda x:1 if 'HDD' in x else 0)
df3['Layer2SSD'] = df3['first'].apply(lambda x:1 if 'SSD' in x else 0)
df3['Layer2Hybrid'] = df3['first'].apply(lambda x:1 if 'Hybrid' in x else 0)
df3['Layer2Flash_Storage'] = df3['first'].apply(lambda x:1 if 'Flash_Storage' in x else 0)

df3['second'] = df3['second'].str.replace(r'\D', '')

df3['first'] = df3['first'].astype(int)
df3['second'] = df3['second'].astype(int)

df3['HDD'] = (df3['first']*df3['Layer1HDD'] + df3['second']*df3['Layer2HDD']) 
df3['SSD'] = (df3['first']*df3['Layer1SSD'] + df3['second']*df3['Layer2SSD']) 
df3['Hybrid'] = (df3['first']*df3['Layer1Hybrid'] + df3['second']*df3['Layer2Hybrid']) 
df3['Flash_Storage'] = (df3['first']*df3['Layer1Flash_Storage'] + df3['second']*df3['Layer2Flash_Storage']) 

df3.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
       'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
       'Layer2Flash_Storage'],inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=2)
```

```python
x_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

# Linear Regrassion

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,10,15,16])
],remainder='passthrough')

step2 = LinearRegression()

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(x_train,y_train)

y_pred = pipe.predict(x_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

# Decision Tree

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,10,15,16])
],remainder='passthrough')

step2 = DecisionTreeRegressor(max_depth=8)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(x_train,y_train)

y_pred = pipe.predict(x_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

# Random Forest

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,10,15,16])
],remainder='passthrough')

step2 = RandomForestRegressor(n_estimators=100,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(x_train,y_train)

y_pred = pipe.predict(x_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```
